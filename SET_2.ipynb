```python
#  Read CSV
import pandas as pd

df = pd.read_csv('SET_2.csv')
df.head()
```

```python
# 1: Create a bar chart comparing news sold in each month
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.bar(df['month'], df['news_sold'], color='GREEN')
plt.title('rAJAN- News Sold per Month')
plt.xlabel('Month')
plt.ylabel('News Sold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 2: Add "percentage_change" column (month-to-month change)
df['percentage_change'] = df['news_sold'].pct_change() * 100
df
```

```python
# 3: Filter months where news_sold < 2000
low_sales = df[df['news_sold'] < 2000]
print("Q3: Months with news_sold < 2000:")
print(low_sales)
```

```python
# 4: Group months into quarters and calculate total news sold per quarter

# Define mapping for quarters
quarter_map = {
    'January': 'Q1', 'February': 'Q1', 'March': 'Q1',
    'April': 'Q2', 'May': 'Q2', 'June': 'Q2',
    'July': 'Q3', 'August': 'Q3', 'September': 'Q3',
    'October': 'Q4', 'November': 'Q4', 'December': 'Q4'
}
df['quarter'] = df['month'].map(quarter_map)

# Group and calculate total news sold per quarter
quarterly_sales = df.groupby('quarter')['news_sold'].sum().reset_index()
print("Q4: Total news sold per quarter:")
print(quarterly_sales)
```

```python
# 5: Save months with above-average news sold to a new CSV
average_news = df['news_sold'].mean()
above_avg = df[df['news_sold'] > average_news]

# FOR Save to CSV
above_avg.to_csv('above_average_news.csv', index=False)

print("Q5: Months with above-average news sold saved to 'above_average_news.csv'")
above_avg
```
