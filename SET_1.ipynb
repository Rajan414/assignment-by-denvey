```python
# Read the CSV into a DataFrame

import pandas as pd

df = pd.read_csv('SET_1.csv')
df.head()
```

```python
# 1: Plot a line graph showing the number of news sold for each month
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(df['month'], df['news_sold'], marker='o', linestyle='-', color='RED')
plt.title('RAJAN-Monthly News Sold')
plt.xlabel('Month')
plt.ylabel('News Sold')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 2: Add a new column 'sales_category' (High if >3000, else Low)
df['sales_category'] = df['news_sold'].apply(lambda x: 'High' if x > 3000 else 'Low')
df
```

```python
# 3: Sort the DataFrame in descending order based on 'news_sold'
df_sorted = df.sort_values(by='news_sold', ascending=False)
df_sorted
```

```python
# 4: Calculate and print the average number of news sold
average_sales = df['news_sold'].mean()
print("Q4: Average number of news sold:", round(average_sales, 2))
```

```python
# 5: Find and display the month(s) with the maximum number of news sold
max_sold = df['news_sold'].max()
max_months = df[df['news_sold'] == max_sold]

print("Q5: Month(s) with the maximum number of news sold:")
print(max_months)
```
